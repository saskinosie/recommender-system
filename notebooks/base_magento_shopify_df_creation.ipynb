### Creating base magento retail and wholesale DFs

```python
mag_items = pd.read_csv("../data/magento_items_base.csv")
mag_items
```

```python
# Creating new DF of just retail orders
magento_retail = mag_items.loc[mag_items['store_id'] == 1]
magento_retail
```

```python
# Creating new DF of just wholesale orders
magento_wholesale = mag_items.loc[mag_items['store_id'] == 2]
magento_wholesale
```

### Creating base shopify retail and wholesale DFs

```python
shop_items = pd.read_csv('../data/shopify_items_base.csv')
shop_items.head()
```

```python
# Renaming columns for congruency
shop_items = shopify_items.rename(columns = {'Name' : 'order_id', 'Lineitem name' : 'name', 'Lineitem quantity' : 'qty_ordered', 'Email' : 'email', 'Tags' : 'store_id'})
shop_items.head()
```

```python
# Creating new DF of just retail orders
shopify_retail = shop_items.loc[shop_items['store_id'] == 1]
shopify_retail
```

```python
# Creating new DF of just wholesale orders
shopify_wholesale = shop_items.loc[shop_items['store_id'] == 2]
shopify_wholesale
```

### Combining base Magento and Shopify base DFs for retail and wholesale

```python
# Combining retail DFs from magento and shopify
base_retail = pd.concat([magento_retail,shopify_retail], axis=0, ignore_index=True)
base_retail
```

```python
# Removing unnecessary text from rows
base_retail['name'] = base_retail['name'].str.replace('™','', regex=True)
```

```python
# Removing unnecessary text from rows
base_retail['name'] = base_retail['name'].str.replace('"','', regex=True)
```

```python
# Removing unnecessary text from rows
base_retail['name'] = base_retail['name'].str.replace('â','', regex=True)
```

```python
# Removing unnecessary text from rows
base_retail['name'] = base_retail['name'].str.replace('¢','', regex=True)
```

```python
# Removing unnecessary text from rows
base_retail['name'] = base_retail['name'].str.replace('â','', regex=True)
```

```python
# Removing unnecessary text from rows
base_retail['name'] = base_retail['name'].str.replace('â¢','', regex=True)
```

```python
# Saving base retail df to csv
base_retail.to_csv('../data/base_retail_no_email.csv', index = False)
```
